```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
```

```python
input_xor = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
label_xor = np.array([[0], [1], [1], [0]], dtype=np.float32)
```

```python
model = Sequential([Dense(10, input_shape=(2,), activation='relu')])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(input_xor, label_xor, epochs=275)
```

```python
model.evaluate(input_xor,label_xor)
```

```python
input_test = np.array([[1,1],[0,1],[0,0],[1,0]])
```

```python
model.predict(input_test)
```

```python
model_multilayer = Sequential([
    Dense(10, input_shape=(2,), activation='relu'),
    Dense(10, activation='relu'),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model_multilayer.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_multilayer.fit(input_xor, label_xor, epochs=275)
```

```python
model.evaluate(input_xor,label_xor)
```

```python
model_multilayer.predict(input_test)
```

```python
model_multilayer_sigmoid = Sequential([
    Dense(10, input_shape=(2,), activation='sigmoid'),
    Dense(10, activation='sigmoid'),
    Dense(10, activation='sigmoid'),
    Dense(1, activation='sigmoid')
])
```

```python
model_multilayer_sigmoid.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_multilayer_sigmoid.fit(input_xor, label_xor, epochs=275)
```

```python
model_multilayer_sigmoid.evaluate(input_test,label_xor)
```

```python
model_multilayer_sigmoid.predict(input_test)
```

```python
model_multilayer_tanh = Sequential([
    Dense(10, input_shape=(2,), activation='tanh'),
    Dense(10, activation='tanh'),
    Dense(10, activation='tanh'),
    Dense(1, activation='sigmoid')
])
```

```python
model_multilayer_tanh.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_multilayer_tanh.fit(input_xor, label_xor, epochs=275)
```

```python
model_multilayer_tanh.evaluate(input_test,label_xor)
```

```python
model_multilayer_tanh.predict(input_test)
```
